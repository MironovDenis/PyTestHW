```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(n=20)
```

```python
one_hot_data = pd.DataFrame({'IAmHuman':lst})
```

```python
one_hot_data.loc[one_hot_data['IAmHuman'] == 'human', 'IAmHuman'] = 1
one_hot_data.loc[one_hot_data['IAmHuman'] == 'robot', 'IAmHuman'] = 0
```
